# Quel est le mois durant lequel nous avons realise le meilleur chiffre d'affaire?

```python
#fonction pour extraire les mois
def month(x):
    return x.split('/')[0]
month('12/30/19 00:01')    
```

```python
all_data['Month']=all_data['Order Date'].apply(month)
all_data
```

```python
all_data['Month'].unique()
```

```python
#supprimer order date
all_data=all_data[all_data['Month']!='Order Date']
all_data['Month'].unique()
```

```python
all_data.dtypes
```

```python
#changer type de Month 
all_data['Month']=all_data['Month'].astype(int)
all_data.dtypes
```

```python
#Pour trouver chiffre d'affaire on va multiplier Price Each,Quantity Ordered et pouir ca il faut changer le type 
all_data['Price Each']=all_data['Price Each'].astype(float)
all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype(int)
all_data.dtypes
```

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data
```

```python
all_data.groupby('Month')['Sales'].sum()
```

```python
months=range(1,13)
plt.figure(figsize=(5, 3))  # Ajuste la taille de la figure

plt.bar(months,all_data.groupby('Month')['Sales'].sum())
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month number')
plt.show()
```

# En quelle moment doit on faire une compagne publicitaire pour avoir plus de ventes ?

```python
all_data
```

```python
all_data['Hour']=pd.to_datetime(all_data['Order Date']).dt.hour
```

```python
all_data
```

```python
keys=[]
hours=[]
for key,hour in all_data.groupby('Hour'):
    keys.append(key)
    hours.append(len(hour))
hours
```

```python
plt.grid()
plt.plot(keys,hours)
plt.xlabel('heure de la journee')
plt.ylabel('nombre des comandes')
```
